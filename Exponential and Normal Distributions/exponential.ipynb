#### Exponential Distribution and Service Time Modeling

```python
import numpy as np

import matplotlib.pyplot as plt

%matplotlib inline
```

```python
np.random.seed(42)  # For reproducibility
slots = 1000
mu = 5
serviceDurations = np.random.exponential(1/mu, slots)
```

```python
plt.figure(figsize=(10, 6)) # Sets the figure size
plt.hist(serviceDurations, bins=30, edgecolor='black')
plt.xlabel("Call duration")
plt.ylabel("Probability Density")
plt.title("Histogram of Exponential Distribution (mu=5)")
plt.grid(axis='y', alpha=0.75) 

# Display the plot
plt.show()
```

#### Normal Distribution and Financial Risk

```python
slots = 1000
mean = 0.0005
SD = 0.015
dailyReturns = np.random.normal(mean, SD, slots)
```

```python
plt.figure(figsize=(10, 6)) # Sets the figure size
plt.hist(dailyReturns, bins=30, edgecolor='black')
plt.xlabel("Daily Returns")
plt.ylabel("Probability Density")
plt.title("Histogram of Normal Distribution (mean=0.0005, SD=0.015)")
plt.grid(axis='y', alpha=0.75) 

# Display the plot
plt.show()
```

Calculating the empirical 1st percentile

```python
empirical_1st_percentile = np.percentile(dailyReturns, 1)
print(f"Empirical 1st Percentile: {empirical_1st_percentile * 100:.2f} %")
```

brief discussion

The theoretical 1st percentile is -3.4% while the empirical 1st percentile is -3.5%. these two values are very close with a small difference of 0.1%.
